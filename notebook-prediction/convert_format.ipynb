```python
reshaped_df = df.set_index('Komoditas (Rp)').T
reshaped_df.index.name = 'Date'
reshaped_df = reshaped_df.iloc[1:].reset_index(drop=True)
reshaped_df
```

```python
numeric_columns = reshaped_df.columns[1:]  # Columns starting from dates

reshaped_df[numeric_columns] = (
    reshaped_df[numeric_columns]
    .replace(',', '', regex=True)
    .astype(int)
)

reshaped_df.to_csv('data/HargaBawangMerah2022.csv', index=False)
```
